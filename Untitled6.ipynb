```python
import time
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns


# Paths to the models
conv2d_model_path = 'cnn_model_tf.h5'
resnet_model_path = 'resnet50_transfer_learning_tf.h5'

# ViT model performance data
vit_results = {
    'eval_loss': 0.09825846552848816,
    'eval_accuracy': 0.9709412507896399,
    'eval_precision': 0.9797235023041475,
    'eval_recall': 0.9779208831646734,
    'eval_f1': 0.9788213627992634,
    'eval_runtime': 117.6481,
    'eval_samples_per_second': 13.455,
    'eval_steps_per_second': 1.683
}

test_data_dir = 'D:/01 STUDY MATERIAL/ai project/Breast-Splitted/test'

# Image data generator
image_size = (224, 224)  # Ensure this matches model input size
batch_size = 32

test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='binary',
    shuffle=False
)

# Function to evaluate model performance
def evaluate_model(model_path, test_generator, model_name='Model'):
    model = load_model(model_path)
    start_time = time.time()
    y_pred = model.predict(test_generator)
    eval_time = time.time() - start_time

    y_pred_classes = np.round(y_pred).astype(int)
    y_true = test_generator.classes

    report = classification_report(y_true, y_pred_classes, target_names=test_generator.class_indices.keys(), output_dict=True)
    confusion = confusion_matrix(y_true, y_pred_classes)

    # Compute ROC curve and AUC
    fpr, tpr, _ = roc_curve(y_true, y_pred)
    roc_auc = auc(fpr, tpr)

    print(f"\n{model_name} Performance:")
    print(f"Evaluation Time: {eval_time:.2f}s")
    print(f"Accuracy: {report['accuracy']:.4f}")
    print(f"Precision: {report.get('1', {}).get('precision', 0):.4f}")
    print(f"Recall: {report.get('1', {}).get('recall', 0):.4f}")
    print(f"F1 Score: {report.get('1', {}).get('f1-score', 0):.4f}")

    # Confusion Matrix
    plt.figure(figsize=(6, 4))
    sns.heatmap(confusion, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix for {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # ROC Curve
    plt.figure(figsize=(6, 4))
    plt.plot(fpr, tpr, color='blue', label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='grey', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve for {model_name}')
    plt.legend(loc='lower right')
    plt.show()

# Evaluate both models
evaluate_model(conv2d_model_path, test_generator, model_name='Conv2D Model')
evaluate_model(resnet_model_path, test_generator, model_name='ResNet Model')

# Display ViT model performance
def display_vit_performance(vit_results):
    print("\nViT Model Performance:")
    print(f"Evaluation Time: {vit_results['eval_runtime']:.2f}s")
    print(f"Accuracy: {vit_results['eval_accuracy']:.4f}")
    print(f"Precision: {vit_results['eval_precision']:.4f}")
    print(f"Recall: {vit_results['eval_recall']:.4f}")
    print(f"F1 Score: {vit_results['eval_f1']:.4f}")

display_vit_performance(vit_results)
```
